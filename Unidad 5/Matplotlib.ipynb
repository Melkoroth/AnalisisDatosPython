```python
#MATPLOTLIB
import numpy as np                         # importación de numpy
import matplotlib as mpl                   # importación de matplotlib con alias mpl
import matplotlib.pyplot as plt
#%matplotlib inline
```

```python
# generación de datos
x = np.linspace(start=0, stop=2*np.pi, num=50)
y1 = np.sin(x)
plt.plot(x, y1);
```

```python
# datos
y1 = np.sin(x)
y2 = np.cos(x)  
# representación : 
fig, (ax1, ax2) = plt.subplots(nrows= 1,ncols = 2, figsize=(20,5))
ax1.plot(x, y1);
#ax2.plot(x, y2); 
```

```python
fig, ax = plt.subplots(nrows = 1, figsize=(10,5))
ax.plot(x, y1)
ax.plot(x, y2);   # Prueba a quitar el ;
```

```python
fig, ax = plt.subplots(1, figsize=(10,5))

# Opción: Pinta una cuadrícula en la gráfica
ax.grid(True)   

ax.plot(x, y1)
ax.plot(x, y2);
```

```python
fig, ax1 = plt.subplots(1, figsize=(10,5))

ax1.grid(True)            

ax1.plot(x, y1)
ax1.plot(x, y2);

# Opción: Aajusta el gráfico a los datos que contiene
x_min, x_max, y_min, y_max = ax1.axis('tight');
ax1.axis((x_min, x_max, y_min, y_max));
```

```python
print(x_min, x_max, y_min, y_max)
```

```python
fig, ax = plt.subplots(1, figsize=(10,5))
ax.grid(True)            

ax.plot(x, y1, label='seno')         # nombre de la línea que representa el seno
ax.plot(x, y2, label='coseno')       # nombre de la línea que representa el coseno

ax.title.set_text("Gráficos con Matplotlib")
ax.legend(loc=0);        
```
